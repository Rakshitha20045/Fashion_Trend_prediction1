```python
def df_to_windowed_df(file_path, start_date_str, end_date_str, n=3):
    # Load CSV file
    df = pd.read_csv(file_path)
    
    # Convert date strings to datetime objects
    start_date = pd.to_datetime(start_date_str)
    end_date = pd.to_datetime(end_date_str)
    
    # Convert crawl_timestamp to datetime if needed
    if isinstance(df['crawl_timestamp'].iloc[0], str):
        df['crawl_timestamp'] = pd.to_datetime(df['crawl_timestamp'])
    
    # Filter dataframe between start_date and end_date
    df_subset = df[(df['crawl_timestamp'] >= start_date) & (df['crawl_timestamp'] <= end_date)]
    
    dates = []
    X, Y = [], []
    
    for i in range(len(df_subset) - n):
        # Extract window of size n + 1
        window = df_subset.iloc[i:i + n + 1]
        
        # Check if the window has exactly n + 1 rows
        if len(window) == n + 1:
            # Extract X (features) and Y (target)
            x = window['variant_price'].values[:-1]
            y = window['variant_price'].values[-1]
            
            # Append to lists
            dates.append(window['crawl_timestamp'].iloc[-1])
            X.append(x)
            Y.append(y)
    
    # Create DataFrame from lists
    ret_df = pd.DataFrame({'Target Date': dates})
    for i in range(n):
        ret_df[f'Target-{n-i}'] = [x[i] for x in X]
    ret_df['Target'] = Y
    
    return ret_df

# Example usage
file_path = 'C:/Users/RAKSHITHA/OneDrive/Documents/Continuous_DateS_Range-LAPTOP-98PL5KBA.csv'
windowed_df = df_to_windowed_df(file_path, '2019-01-01', '2024-12-31', n=3)
print(windowed_df)
```

```python
def windowed_df_to_date_X_y(windowed_dataframe):
  df_as_np = windowed_dataframe.to_numpy()

  dates = df_as_np[:, 0]

  middle_matrix = df_as_np[:, 1:-1]
  X = middle_matrix.reshape((len(dates), middle_matrix.shape[1], 1))

  Y = df_as_np[:, -1]

  return dates, X.astype(np.float32), Y.astype(np.float32)

dates, X, y = windowed_df_to_date_X_y(windowed_df)

dates.shape, X.shape, y.shape
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def evaluate_model_performance(y_true, y_pred):
    # Calculate MSE, MAE, and R-squared
    mse = mean_squared_error(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    
    # Print the metrics
    print(f"Mean Squared Error (MSE): {mse}")
    print(f"Mean Absolute Error (MAE): {mae}")
    print(f"R-squared (R2): {r2}")
    
    # Return metrics as a dictionary if needed
    return {'mse': mse, 'mae': mae, 'r2': r2}

# Example usage:
# Assuming you have predictions for train, validation, and test sets: train_predictions_enhanced, val_predictions_enhanced, test_predictions_enhanced
print("Training Set Performance:")
evaluate_model_performance(y_train, train_predictions_enhanced)

print("\nValidation Set Performance:")
evaluate_model_performance(y_val, val_predictions_enhanced)

print("\nTest Set Performance:")
evaluate_model_performance(y_test, test_predictions_enhanced)

```
